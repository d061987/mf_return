```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
URL = 'https://www.amfiindia.com/nav-history-download'


Source = requests.get(URL)
soup = bs(Source.text, 'html.parser')
```

```python
NavDownMFName = []
for table in soup.find_all('select',{'class': 'select', 'id': 'NavDownMFName'})[0]:
    str_table = str(table)[15:].split('">')
    if len(str_table)>1:
        mf_Id = str_table[0]
        mf_Name = str_table[1].split('<')[0]
        NavDownMFName.append(mf_Id + ' ' +mf_Name)

NavDownMFName
```

```python
NavType = []

for table in soup.find_all('select', {'class' : 'select', 'id' : 'NavDownType'})[0]:
    str_table = str(table)[15:].split('">')
    if len(str_table) > 1 and str_table[0] != '':
        NV_Value = str_table[0]
        NV_Name = str_table[1].split('<')[0]
        print(NV_Value, NV_Name)
        NavType.append(NV_Value + '  ' + NV_Name)
NavType
```
